```python
import pandas as pd
import re
```

Parsing the original dataset

```python
texts = []
labels = []

with open("movie.csv", encoding="ISO-8859-1") as f:
    next(f)

    for line in f:
        line = line.strip()
        line = line.rstrip("\t").strip('"')
        if "," in line:
            text, label = line.rsplit(",", 1)

            texts.append(text.strip())
            labels.append(label.strip())

df = pd.DataFrame({
    "text": texts,
    "label": labels
})

df = df[df["label"].isin(['0', '1'])]
df = df.dropna(subset=["label"])
df["label"] = df["label"].astype(int)

print(df["label"].value_counts())
print(df.shape)
```

Cleaning HTLM tags, punctuation, extra spaces

```python
def clean_text(text):
    text = text.lower()
    # remove HTML tags if any
    text = re.sub(r"<.*?>", " ", text)
    # remove punctuation except apostrophe
    text = re.sub(r"[^\w\s']", " ", text)
    # remove extra spaces
    text = re.sub(r"\s+", " ", text)
    return text.strip()

df["text"] = df["text"].astype(str).apply(clean_text)

print(type(df["text"].iloc[0]))
print(df["text"].iloc[0])
```

Removing stop words

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

df["text"] = df["text"].apply(
    lambda x: [word for word in x.split() if word not in stop_words]
).apply(
    lambda x: " ".join(x) 
)
```

```python
df.to_csv("movies_clean.csv")
```
